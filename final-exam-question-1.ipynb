### The potential energy

$$ \Delta U = \frac{1}{2} K_1 Q_1^2 + \frac{1}{2} K_2 Q_2^2 + \frac{1}{3} C Q_1 Q_2^2 $$

### Constrained parameters and initial conditions

Take $K_1 = 1$, $K_2 = 0.2$, $M_1 = M_2 = 1$, and $C=0.5$. For the initial conditions, take $Q_1(0) = 0$, $Q_2(0) = 1$, and both velocities starting at zero.

With your code complete, you can test four cases for the initial conditions (all with the initial velocity set to zero):

Why do the dynamics differ when only $Q_1$ is initially displaced compared to when only $Q_2$ is initially displaced? How does this relate to the coupling term in the potential energy? What is the physical interpretation? You could use our cylindrical beam example from class to describe this, if you like.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Parameters
K1 = 1.0
K2 = 0.2
C = 0.5
M1 = M2 = 1.0

# Define the potential energy function
def potential_energy(Q1, Q2):
    return 0.5 * K1 * Q1**2 + 0.5 * K2 * Q2**2 + (1/3) * C * Q1 * Q2**2

# Equations of motion
def equations(t, y):
    Q1, Q2, P1, P2 = y
    dQ1_dt = P1 / M1
    dQ2_dt = P2 / M2
    dP1_dt = -K1 * Q1 - (1/3) * C * Q2**2
    dP2_dt = -K2 * Q2 - C * Q1 * Q2
    return [dQ1_dt, dQ2_dt, dP1_dt, dP2_dt]

# Initial conditions
Q1_0 = 0
Q2_0 = 1
P1_0 = 0
P2_0 = 0
initial_conditions = [Q1_0, Q2_0, P1_0, P2_0]

# Time mesh
t_mesh = np.linspace(0, 50, 500)

# Solve the equations of motion
solution = solve_ivp(equations, [t_mesh[0], t_mesh[-1]], initial_conditions, t_eval=t_mesh)
Q1, Q2 = solution.y[0], solution.y[1]

# Create the grid for the potential energy
Q1_grid, Q2_grid = np.meshgrid(np.linspace(-1.5, 1.5, 100), np.linspace(-1.5, 1.5, 100))
U_grid = potential_energy(Q1_grid, Q2_grid)

# Plotting
plt.figure(figsize=(10, 8))

# Contour plot
plt.contourf(Q1_grid, Q2_grid, U_grid, levels=50, cmap='viridis')
plt.contour(Q1_grid, Q2_grid, U_grid, levels=50, colors='white', linewidths=1)

# Trajectory plot
plt.plot(Q1, Q2, color='red', label="Trajectory")

# Create contour plot
contour_filled = plt.contourf(Q1_grid, Q2_grid, U_grid, levels=50, cmap='viridis')

# Add color bar using the contour plot reference
plt.colorbar(contour_filled, label="Potential Energy")


# Labels and legend
plt.xlabel('$Q_1$', fontsize=14)
plt.ylabel('$Q_2$', fontsize=14)
plt.title('Potential Energy Contour and Trajectory of $Q_1$ and $Q_2$', fontsize=16)
plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)
plt.legend(fontsize=12)
#plt.colorbar(label='Potential Energy')
plt.grid()
plt.show()
```

When $Q_1$ is initially displaced, the trajectory involves coupling through the term $C Q_1 Q_2^2$.
This results in energy transfer between the modes, as the coupling term generates forces proportional to $Q_1$ and $Q_2^2$.
When only $Q_2$ is initially displaced, the dynamics involve direct oscillations driven by $K_2$ and modified by the coupling term.
This coupling reflects the anharmonic interaction, leading to trajectories that differ based on initial conditions.

In physical terms, the system exhibits coupled oscillations where energy is exchanged between the two modes due to the anharmonic term.
This is analogous to our cylindrical beam example, where bending in one direction influences motion in the other due to geometric coupling.
